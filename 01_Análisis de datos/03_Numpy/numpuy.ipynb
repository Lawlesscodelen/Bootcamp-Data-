```python
def m(area=5):
    Z = np.zeros((area,area), [('x',float),('y',float)])
    Z['x'], Z['y'] = np.meshgrid(np.linspace(0,1,area),
                                 np.linspace(0,1,area))
    return Z
```

```python
# Moving Average
def moving_average(a, window=5):
        
    ret = np.cumsum(a)
    ret[window: ] = ret[window: ] - ret[: -window]
    return ret[window-1:] / window

Z = np.random.randn(100)
print(Z)
print(moving_average(Z, 50))
```
